# Part 1: Regular Expressions

## 1. Extracting the name into a vector
We have faced two cases while solving and we would like to mention it. Case1 Gives the answers but there are some limitation in it.

Case1: We only need the range of characters with both case(Upper and Lower) of letter with other characters like dot and comma. Whitespace is also required in it. However, when using "\s" it not only includes spaces between the first and last name but also includes the spaces between the names of two person. We can see the spaces in the vector of the case1.

Case2: It removes the limitation of case1.

```python
#Case1 --> Limitations
string = """555-1239Dr. Anthony Fauci(636) 555-0113Hollingdorp, Donnatella555-6542Fitzgerald, F. Scott555 8904Rev. Martin Luther King636-555-3226Snodgrass, Theodore5553642Carlamina Scarfoni"""
extractedName = re.findall('[a-zA-Z\s.,]+',string)
print("Case1 Vector:")
print(extractedName)

print("Case2 Vector:")
#Case2 --> Proper Answer
string = """555-1239Dr. Anthony Fauci(636) 555-0113Hollingdorp, Donnatella555-6542Fitzgerald, F. Scott555 8904Rev. Martin Luther King636-555-3226Snodgrass, Theodore5553642Carlamina Scarfoni"""
extractedName = re.findall('[a-zA-Z](?:[a-zA-Z., ]*[a-zA-Z])',string)
print(extractedName)
```

## 2. Use the new vector

### a. First Task:  Rearrange the vector to "Firstname and Lastname"  preserving titles or middle/second names. 

```python
#2a
regexp = "([A-Za-z]+) ([A-Za-z]+)";
pattern = re.compile(regexp)
for i,name in enumerate(extractedName):
    '''
    This function uses the new vector "extractedName" to update it to conform to the standard. Initially, it checks the 
    format of the name, if it matches then it keeps is as same, otherwise if the result is none then it checks the patterns
    in the name. Here we have namePattern1 and namePattern2. With respect to their patterns, the names are properly 
    formatted and updated in the same vector. 
    '''
    result = re.match(pattern, name)
    if(result == None):
        namePattern1 = re.search('(\w+), (\w+). (\w+)',name)
        namePattern2 = re.search('((\w+), (\w+))',name)
        if(namePattern1):
            updatedName= re.sub(r'(\w+), (\w+). (\w+)', r'\2. \3 \1',name)
            extractedName[i] = updatedName
        elif(namePattern2):
            updatedName= re.sub(r'(\w+), (\w+)', r'\2 \1',name)
            extractedName[i] = updatedName
extractedName
```

### b. Second task: Names with Title.

```python
#2b
nameWithTitle = []
for name in extractedName:
    '''
    This function uses the updated vector "extractedName" and search if it has title or not, one by one in a loop. 
    If matched, then append to the new vector "nameWithTitle".
    '''
    title = re.search('(\w+). (\w+) (\w+)',name)
    if(title != None):
        nameWithTitle.append(name)
nameWithTitle
```

### c. Third Task: Character with middle/second name

```python

middleName = []

for name in extractedName:
    '''
    Similar to second task, his function uses the updated vector "extractedName" and search if it has middle name or not.
     If matched, then append to the new vector "middleName".
    '''
    middle = re.search('(\w+). (\w+) (\w+) (\w+)',name)
    if(middle != None):
        middleName.append(name)
middleName
```

###  3. HTML String

Here in 3, we are trying to extract title opening tag from the given HTML string <!-- "<title>+++BREAKING NEWS+++<title>" -->. And the suggested regular expression for this is "<.+>". This regex will match the whole given html string <!-- "<title>+++BREAKING NEWS+++<title>" --> because of the +. The reason is that the plus is greedy, it match as much text as possible. That is, the plus causes the regex engine to repeat the preceding token as often as possible. Only if that causes the entire regex to fail, will the regex engine backtrack. That is, it will go back to the plus, make it give up the last iteration, and proceed with the remainder of the regex.
    
As we already know  the first place where it will match is the first "<" in the string. The dot,next token,  this matches any character except a newline. The dot is repeated by the plus. Since the plus is greedy, it will repeat the dot as many times as it can. The dot match all the characters "t","i","t","l","e". We can see the main problem now. The dot also matches ">" and then it continues to repeat. The dot again will match the remaining character of the strings. At the end of the string, ">" can matches. This is how, it is reported that <!-- "<title>+++BREAKING NEWS+++<title>" --> has been successfully matched.
    
Solution: Laziness instead of greediness
  The quick fix is make the plus lazy instead of greedy. Lazy quantifiers are sometimes also called “ungreedy” or “reluctant”. You can do that by putting a question mark after the plus in the regex. You can do the same with the star, the curly braces and the question mark itself. So our regex becomes "<.+?>". Lets review this process.

Again, "<" matches the first "<" in the string. The next token is the dot, this time repeated by a lazy plus. This tells the regex engine to repeat the dot as few times as possible. The minimum is one. So the engine matches the dot with "t". The requirement has been met, and the engine continues with > and then i. This fails.  Again, the engine will backtrack. But this time, the backtracking will force the lazy plus to expand rather than reduce its reach. So the match of .+ is expanded to "ti", and the engine tries again to continue with > and then anoter "t". Again it fails. This process repeats itself until "<title" and then continues with ">". Now, > is matched successfully. The last token in the regex has been matched. The engine reports that title opening tag has been successfully matched.
    
https://www.regular-expressions.info/repeat.html#lazy


```python
#Solution: Laziness instead of greediness. "?"
text = "<title>+++BREAKING NEWS+++</title>"
pattern = re.search("<.+?>",text)
print(pattern)
```

### 4. Binomial theorem

Here in 4, the regex is within big bracket which represents a set. In this set, the first expression is Caret(^). 
This matches the start of the string. But when its inside a set, it means that it matches any character except inside the set. So, it returned two '-'. Therefore, the caret fails to match the given string with the regex. This is one of the reasons that this "[^0-9=+*()]+" fails. So the caret should be removed.

Other Characters like "0-9","=","+","*","()" will help to match the string after the removal of the Caret. The set  "[]" is followed by the plus sign means it matches as much text as possible. However, for the minus sign, there is not any character inside the set to match. If '-' is escaped (e.g. [a\-z]) or if it’s placed as the first or last character (e.g. [-a] or [a-]), it will match a literal '-'.

https://docs.python.org/3/library/re.html

```python
#4) Solution: The cadet inside the set is removed and a minus sign is added inside the set to match it with a literal'-'.
text = "(5-3)**2=5**2-2*5*3+3**2"

pattern = re.findall("[0-9=+*()\-]+",text)
print(pattern)
```

## 1.Read and Store Data

```python
f = open(r"Project1.txt","r")
file = f.read()
list = [file]
```

## 2.Get the Upper Line
### RE
~~~
\s+\d+\s\|\s(?P<Player_name>.+)\|
(?P<total_points>[0-9]+[.][0-9])\s+\|
(?P<R1_R>\w)\s+(?P<R1_O>\d{1,})?\|
(?P<R2_R>\w)\s+(?P<R2_O>\d{1,})?\|
(?P<R3_R>\w)\s+(?P<R3_O>\d{1,})?\|
(?P<R4_R>\w)\s+(?P<R4_O>\d{1,})?\|
(?P<R5_R>\w)\s+(?P<R5_O>\d{1,})?\|
(?P<R6_R>\w)\s+(?P<R6_O>\d{1,})?\|
(?P<R7_R>\w)\s+(?P<R7_O>\d{1,})?\|
~~~
### EXAMPLE
~~~
   28 | SOFIA ADINA STANESCU-BELLU      |3.5  |W  24|D   4|W  22|D  19|L  20|L   8|D  36|  
   62 | ASHWIN BALAJI                   |1.0  |W  55|U    |U    |U    |U    |U    |U    |
~~~
  
### Description
We can see in first example the name include a "-". So we use ".+" to match name, it will stop at "|".  
In some cases, players did not participate in the competition, so they didn't have an opponent, use "\s+(?P<R1_O>\d{1,})?" to match opponent number， it can match spaces or opponent number. 

### Output
We will get a list which include dicts 
~~~
[{},{}.....{}]
~~~

```python
line1 = re.compile("\s+\d+\s\|\s(?P<Player_name>.+)\|(?P<total_points>[0-9]+[.][0-9])\s+\|(?P<R1_R>\w)\s+(?P<R1_O>\d{1,})?\|(?P<R2_R>\w)\s+(?P<R2_O>\d{1,})?\|(?P<R3_R>\w)\s+(?P<R3_O>\d{1,})?\|(?P<R4_R>\w)\s+(?P<R4_O>\d{1,})?\|(?P<R5_R>\w)\s+(?P<R5_O>\d{1,})?\|(?P<R6_R>\w)\s+(?P<R6_O>\d{1,})?\|(?P<R7_R>\w)\s+(?P<R7_O>\d{1,})?\|")
line1_result = [m.groupdict() for m in line1.finditer(list[0])]
print("How many players are matched =", len(line1_result))
line1_result
```

### What does PXX means？
What does Pxx (eg. P3 or P15) in the rating mean?  A P in the rating means "provisional".  The USCF gives you a provisional rating until you have played 25 rated games.  In layman's terms what this means is that the rating can rise or fall dramatically, until the person has played 25 rated games.  In some cases, players who have very high provisional ratings get the rating because they have played other provisional-rated players.  Also, a person who has played fewer than 4 rated games is still considered to be unrated, and as such, should be paired as an unrated.  This comes into play only in Quads tournaments, where pairings are based on one's rating.

reference:http://weibelchess.blogspot.com/2012/10/some-terminology-for-new-players.html

## 3. Get the Lower Line
### RE
~~~
\s+(?P<state>[A-Z]+)\s\|\s\d+\s\/\sR\:\s  
(?P<pre_rating>\d+)(P\d+)?(\s+)?->\d+(P\d+)?
~~~
### EXAMPLE
~~~
ON | 15771592 / R:  955P11-> 979P18  |     |B    |W    |B    |W    |B    |W    |B    |  
MI | 15490981 / R:  377P3 ->1076P10  |     |B    |W    |B    |W    |B    |W    |W    |
MI | 14579262 / R:  967   -> 984     |     |W    |B    |B    |W    |B    |     |     |
~~~
### DESCRIPTION
Use "(?P<state>[A-Z]{2})" to match the state.  
Use "(?P<pre_rating>\d+)" to match the pre rating.  
Whether there are spaces before and after the arrow needs to be judged

```python
line2 = re.compile("\s+(?P<state>[A-Z]{2})\s\|\s\d+\s\/\sR\:\s+(?P<pre_rating>\d+)(P\d+)?\s*->\s*\d+(P\d+)?")
line2_result = [m.groupdict() for m in line2.finditer(list[0])]
print("How many players are matched =",len(line2_result))
line2_result
```

## 4. Convert and merge to dataframe
Convert the list which include dicts to a dataframe object
Then merge the two dataframes horizontally

```python
df_result1 = pd.DataFrame(line1_result)
df_result2 = pd.DataFrame(line2_result)
df_mix = pd.concat([df_result1,df_result2,pd.DataFrame(columns=['pre_ave_opponents_rating'])],axis = 1)
print(df_mix)
```

## 5.Calculate the average pre-tournament rating
Use the data in dataframe to calculate the average pre-tournament rating

```python
round_opponent_column_index = [3,5,7,9,11,13,15]                        #the column index of every round_result in df_mix.
for i in range(df_mix.shape[0]):                                        #df_mix.shape[0] will return the len of df_mix, it is 64
    sum_pre_opponents_rating = 0                                        # use it to store sum ratings
    count = 0                                                           # use it to store how many opponents, because some people didn't take part in full 7 rounds
    for p in round_opponent_column_index:                               # find every opponents' rating
        if (df_mix.iloc[i,p] != None):
            sum_pre_opponents_rating += int(df_mix.iloc[int(df_mix.iloc[i,p])-1,17])
            count += 1
    pre_ave_opponents_rating = int(sum_pre_opponents_rating / count)    #calculate average rating
    df_mix.at[i,'pre_ave_opponents_rating'] = pre_ave_opponents_rating  # store it in df_mix
print(df_mix['pre_ave_opponents_rating'])
```
